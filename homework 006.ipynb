## Задание 2 (необязательное)
Для каждого user_id в файле visit_log.csv определите третий столбец с категорией покупки. Если покупка была, сам файл visit_log.csv изменять не надо.

Запишите в файл funnel.csv визиты из файла visit_log.csv, в которых были покупки с указанием категории.

Учтите условия для данных:

- содержимое purchase_log.txt помещается в оперативную память компьютера;
- содержимое visit_log.csv — нет, используйте только построчную обработку этого файла.

Примеры работы программы при выводе первых трёх строк файла funnel.csv:

```
user_id,source,category
1840e0b9d4,other,Продукты
4e4f90fcfb,context,Электроника
```

````python
import json
import csv


def get_purchase_log():
    output = dict()
    with open ('purchase_log.txt', 'r', encoding='utf8') as fp:
        for idx, row in enumerate(fp):
            row_data = json.loads(row)
            user_id = row_data['user_id']
            if user_id == 'user_id':
                continue
            category = row_data['category']
            if user_id not in output:
                output[user_id] = []
            output[user_id].append(category)
    return output


if __name__ == '__main__':
    print('Задание 1', end='\n\n')
    
    user_categories = get_purchase_log()
    c = 0
    for key, categories in user_categories.items():
#        if len(categories) == 1:
#            continue
        print('{} ‘{}‘'.format(key, ', '.join(categories)))
        c += 1
        if c > 1:
            break

    print()
    print('Задание 2', end='\n\n')

    with open('visit_log.csv', 'r', encoding='utf8') as fp_in:
        with open('funnel.csv', 'w', encoding='utf8', newline='') as fp_out:
            reader = csv.reader(fp_in, dialect='excel')
            writer = csv.writer(fp_out, dialect='excel')
            for idx, row in enumerate(reader):
                if idx == 0:
                    row.append('category')
                else:
                    user_id = row[0]
                    if user_id in user_categories:
                        row.append('|'.join(user_categories[user_id]))
                    else:
                        continue
                writer.writerow(row)

    with open('funnel.csv', 'r', encoding='utf8') as fp:
        for idx, row in enumerate(fp):
            print(row, end='')
            if idx > 1:
                break
````
