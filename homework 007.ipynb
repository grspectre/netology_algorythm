# Параллельные вычисления

## Задание 1
Требуется параллельно (насколько возможно с помощью потоков) выполнить вычисления по формуле 1, вычисления по формуле 2. Затем выполнить вычисления по формуле 3, которые выглядят следующим образом: результат вычислений 1 + результат вычислений 2.

Выполнить последовательно на 10 000 итераций и 100 000 итераций.
```
Формула 1: f(x) = x^2-x^2+x4-x5+x+x
Формула 2: f(x) = x+x
```

Вывести длительность выполнения всех 10 000 итераций и 100 000 итераций в секундах в разбивке по шагам вычислений 1, 2 и 3.

## Задание 2 (необязательное)
Требуется параллельно (насколько возможно с помощью процессов) выполнить вычисления по формуле 1, вычисления по формуле 2. Затем выполнить вычисления по формуле 3, которые выглядят следующим образом: результат вычислений 1 + результат вычислений 2.

Выполнить последовательно на 10 000 итераций и 100 000 итераций.
```
Формула 1: f(x) = x^2-x^2+x4-x5+x+x
Формула 2: f(x) = x+x
```

Вывести длительность выполнения всех 10 000 итераций и 100 000 итераций в секундах в разбивке по шагам вычислений 1, 2 и 3.

````python
from time import sleep, perf_counter
import threading

class Data:

    data = {}

    @classmethod
    def add_data(cls, g_key, key, value):
        if g_key not in cls.data:
            cls.data[g_key] = {}
        cls.data[g_key][key] = value
        
    @classmethod
    def get(cls, g_key, key):
        try:
            return cls.data[g_key][key]
        except:
            return None


def f1(x):
    Data.add_data('f1', x, x**2 - x**2 + x*4 - x*5 + x + x)


def f2(x):
    Data.add_data('f2', x, x + x)


def f3(x):
    f1 = Data.get('f1', x)
    f2 = Data.get('f2', x)
    Data.add_data('f3', x, f1 + f2)


def main(max_value):
    start = perf_counter()

    f1t = [threading.Thread(target=f1, args=[i]) for i in range(max_value)]
    for thread in f1t:
        thread.start()
    for thread in f1t:
        thread.join()
    step_one = perf_counter()

    f2t = [threading.Thread(target=f2, args=[i]) for i in range(max_value)]
    for thread in f2t:
        thread.start()
    for thread in f2t:
        thread.join()
    step_two = perf_counter()

    f3t = [threading.Thread(target=f3, args=[i]) for i in range(max_value)]
    for thread in f3t:
        thread.start()
    for thread in f3t:
        thread.join()

    finish = perf_counter()
    step1 = step_one - start
    step2 = step_two - step_one
    step3 = finish - step_two
    all_step = finish - start
    print("Выполнение {} step1: {} step2: {} step3: {} all: {}.".format(max_value, step1, step2, step3, all_step))


if __name__ == '__main__':
    main(10000)
    main(100000)
````
