# Задание

Написать код на Python в среде Jupyter Notebook для решения следующей задачи.

Вам нужно помочь секретарю автоматизировать работу. Для этого нужно написать программу, которая будет на основе хранимых данных исполнять пользовательские команды.

Исходные данные имеют следующую структуру:

```
1. перечень всех документов
documents = [
{‘type’: ‘passport’, ‘number’: ‘2207 876234’, ‘name’: ‘Василий Гупкин’},
{‘type’: ‘invoice’, ‘number’: ‘11-2’, ‘name’: ‘Геннадий Покемонов’},
{‘type’: ‘insurance’, ‘number’: ‘10006’, ‘name’: ‘Аристарх Павлов’}
]

2. перечень полок, на которых хранятся документы (если документ есть в documents, он обязательно должен быть и в directories)
directories = {
‘1’: [‘2207 876234’, ‘11-2’],
‘2’: [‘10006’],
‘3’: []
}
```

## Общие требования к программе:

- код должен быть грамотно декомпозирован: каждая функция отвечает за свою конкретную задачу, дублирующиеся функции переиспользуются, а их код не повторяется;
- в коде отсутствуют глобальные переменные, за исключением documents и directories;
- пользовательский ввод обрабатывается в цикле while до тех пор, пока пользователь явно не завершит программу вводом команды q.

## Задача 1

Пользователь по команде p может узнать владельца документа по его номеру.

Пример работы программы:

```
Введите команду:
p

Введите номер документа:
10006

Результат:
Владелец документа: Аристарх Павлов
```

## Задача 2 (необязательная)

Секретарь по команде s может по номеру документа узнать, на какой полке документ хранится.

Пример работы программы:
```
Введите команду:
s

Введите номер документа:
10006

Результат:
Документ хранится на полке: 2
```

````python
documents = [
    {"type": "passport", "number": "2207 876234", "name": "Василий Гупкин"},
    {"type": "invoice", "number": "11-2", "name": "Геннадий Покемонов"},
    {"type": "insurance", "number": "10006", "name": "Аристарх Павлов"}
]

directories = {
    "1": ["2207 876234", "11-2"],
    "2": ["10006"],
    "3": []
}


def print_owners(document_number):
    global documents
    response = [i["name"] for i in documents if i["number"] == document_number]
    if len(response) == 0:
        print("Ничего не найдено.")
    for item in response:
        print("Владелец документа: {}".format(item))
    print()


def print_directories(document_number):
    global directories
    response = [shelf for (shelf, doc_numbers) in directories.items() if document_number in doc_numbers]
    if len(response) == 0:
        print("Ничего не найдено.")
    for item in response:
        print("Документ хранится на полке: {}".format(item))
    print()


def get_document_number():
    while True:
        number = input("Введите номер документа:")
        print()
        number = number.strip()
        if number == "":
            continue
        return number


def execute_command(symbol):
    doc_number = get_document_number()
    print("Результат:")
    if symbol == 'p':
        print_owners(doc_number)
    if symbol == 's':
        print_directories(doc_number)


def command():
    symbol = input("Введите команду:")
    available_commands = ["p", "s", "q"]
    
    if symbol not in available_commands:
        print('Команда не распознана. Доступны следующие команды:', ', '.join(available_commands))
        return True
    if symbol == "q":
        return False
    execute_command(symbol)


def main():
    while True:
        is_exec = command()
        if is_exec == False:
            print('Выполнение программы остановлено.')
            break


main()
````
