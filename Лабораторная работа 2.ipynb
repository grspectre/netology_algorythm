Задание: написать программу, которая поможет пользователям определить, сколько денег они могут сэкономить на покупках в разных магазинах.

## Условия:

- Пользователь должен ввести список товаров, которые он собирается купить, и цены на эти товары в разных магазинах.
- Затем программа должна вывести список магазинов и общую стоимость покупок в каждом магазине, а также сообщить, в каком магазине пользователь может сэкономить больше всего денег.
- Список товаров и магазинов определяется самостоятельно.

Выполненное задание разместите на GitHub.

```python
from typing import List
import json
import random

def generate_list(products: List) -> List:
    """
        Генерирует список цен в 3 магазинах
    """
    stores = ['Магазин 1', 'Магазин 2', 'Магазин 3']
    for product in products:
        line = []
        line.append(product)
        for store in stores:
            line.append(store)
            line.append(str(random.randint(100, 100000)/100))
        print(';'.join(line))

        
products = ['Молоко', 'Хлеб', 'Сметана', 'Вафли', 'Зефир глазированный', 'Сахар', 'Чипсы фермерские']
generate_list(products)
```

```python
input = '''Молоко;Магазин 1;432.06;Магазин 2;82.63;Магазин 3;229.4
Хлеб;Магазин 1;788.29;Магазин 2;45.39;Магазин 3;257.72
Сметана;Магазин 1;250.6;Магазин 2;675.86;Магазин 3;242.41
Вафли;Магазин 1;376.23;Магазин 2;18.03;Магазин 3;973.52
Зефир глазированный;Магазин 1;431.21;Магазин 2;466.01;Магазин 3;638.85
Сахар;Магазин 1;754.76;Магазин 2;979.68;Магазин 3;251.6
Чипсы фермерские;Магазин 1;168.72;Магазин 2;600.5;Магазин 3;741.29'''

input = input.split("\n")
data = {}

# для красивого вывода
max_len_of_product = 0

# собираем данные по магазинам
for line in input:
    line = line.split(';')
    product = line.pop(0)

    # Вычисляем самое длинное название у продукта
    if len(product) > max_len_of_product:
        max_len_of_product = len(product)

    for i in range(3):
        store = line.pop(0)
        price = float(line.pop(0))
        if store not in data:
            data[store] = {}
        data[store][product] = price

# Вычисляем дешёвый магазин
result = None
store_name = None
for store in data:
    summary = 0
    for product in data[store]:
        summary += data[store][product]
    if result is None or summary < result:
        result = summary
        store_name = store
        
template = '{:>'+str(max_len_of_product)+'} | {:7.2f}'

# Выводим результат
for store in data:
    summary = 0
    print(store if store_name != store else '{} (Лучшая цена)'.format(store))
    print('-'*(max_len_of_product + 10))
    for product in data[store]:
        price = data[store][product]
        summary += price
        ws_len = max_len_of_product - len(product)
        print(template.format(product, price))
    print('-'*(max_len_of_product + 10))
    print(template.format('Итого', summary))
    print()
```
