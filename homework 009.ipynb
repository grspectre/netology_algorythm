### Задание 2.

```
Написанная функция удаляет все последовательные повторы слов из заданной строки при помощи регулярных выражений.
Пример работы программы:
some_string = ‘Напишите функцию функцию, которая будет будет будет будет удалять все все все все последовательные повторы слов из из из из заданной строки строки при помощи регулярных выражений’

Результат: Напишите функцию, которая будет удалять все последовательные повторы слов из заданной строки при помощи регулярных выражений
```

````python
import re


def check_auto_number(number: str) -> bool:
    pattern = '^[АВЕКМНОРСТУХABEKMHOPCTYX][\d]{3}[АВЕКМНОРСТУХABEKMHOPCTYX]{2}\d{2,3}$'
    return re.match(pattern, number) is not None


def get_check_message(number: str) -> str:
    if check_auto_number(number):
        return "Номер {} валиден. Регион: {}".format(number[0:6], number[6:])
    else:
        return "Номер не валиден."


def main() -> None:
    print('Задание 1')
    numbers = ['А222BС96', 'АБ22ВВ193']
    for number in numbers:
        print("car_id = '{}'".format(number))
        print("Результат: {}".format(get_check_message(number)))
        print()

    print('Задание 2')
    text = 'Напишите функцию функцию, которая будет будет будет будет удалять все все все все последовательные повторы слов из из из из заданной строки строки при помощи регулярных выражений'
    words = re.findall('\w+', text)

    last_word = None
    words_list = []
    for word in words:
        if last_word == word:
            if len(words_list) == 0 or words_list[-1][0] != word:
                words_list.append([word, 1])
            else:
                words_list[-1][1] += 1
        last_word = word
    for word, count in words_list:
        replace = "{}{}".format(word, (' '+word)*count)
        text = text.replace(replace, word)
    print(text)


if __name__ == '__main__':
    main()
````
