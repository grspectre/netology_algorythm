```python
import csv
# инициализируем файл
data = [
    {'name': 'Иван', 'age': 30, 'occupation': 'учитель', 'city': 'Санкт-Петербург'},
    {'name': 'Максим', 'age': 27, 'occupation': 'программист', 'city': 'Екатеринбург'},
    {'name': 'Елена', 'age': 28, 'occupation': 'врач', 'city': 'Новосибирск'},
    {'name': 'Анна', 'age': 43, 'occupation': 'инженер', 'city': 'Москва'},
    {'name': 'Ольга', 'age': 25, 'occupation': 'дизайнер', 'city': 'Краснодар'},
]
with open('person.csv', 'w', newline='', encoding='utf8') as fp:
    writer = csv.writer(fp, delimiter=';', quotechar='"')
    if len(data) > 0:
        is_titles_saved = False
        for row in data:
            if not is_titles_saved:
                keys = data[0].keys()
                writer.writerow(keys)
                is_titles_saved = True
            item = []
            for key in keys:
                item.append(row[key])
            writer.writerow(item)
```

```python
import os
import csv
import yaml
from typing import List

def read_csv(filename: str) -> List:
    full_filename = '{}.csv'.format(filename)
    if not os.path.exists(full_filename):
        raise RuntimeError('File {} does not exist'.format(full_filename))
    with open(full_filename, 'r', encoding='utf8') as fp:
        reader = csv.reader(fp, delimiter=';', quotechar='"')
        keys = None
        output = []
        for row in reader:
            if keys is None:
                keys = row
                continue
            item = {}
            for idx, key in enumerate(keys):
                item[key] = row[idx]
            output.append(item)
        return output


def write_yaml(filename: str, data) -> None:
    full_filename = '{}.yaml'.format(filename)
    with open(full_filename, 'w', encoding='utf8') as fp:
        yaml.dump(data, fp, allow_unicode=True)


def transform(filename: str, from_format: str, to_format: str) -> None:
    available_transformers = ['csv', 'yaml']
    if from_format not in available_transformers:
        print('{} is not available: {}'.format(from_format, ', '.join(available_transformers)))
    if to_format not in available_transformers:
        print('{} is not available: {}'.format(to_format, ', '.join(available_transformers)))
    data = None
    if from_format == 'csv':
        data = read_csv(filename)
    
    if data is None:
        raise RuntimeError('Data is None. Can not get data from {} with name "{}"'.format(from_format, filename))
    
    if to_format == 'yaml':
        write_yaml(filename, data)

        
if __name__ == '__main__':
    transform('person', 'csv', 'yaml')
```
